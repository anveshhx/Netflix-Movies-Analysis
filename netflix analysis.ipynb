categorizing Vote_Average Column
i would cut the Vote_Average column and replace it with: popular, average, below_avg, least_popular
i will use categorize_col() function

```python
df.head()
```

```python
def categorize_col(df, col, labels):

    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'], 
             df[col].describe()['75%'],
             df[col].describe()['max']]

    df[col] = pd.cut(df[col], edges, labels = labels, duplicates = 'drop')
    return df
                  
              
```

used col as arguement in the user definedd function so that infututre i need to change another column we can do that without writing from scratch
col instead of vote average is just the entire columns out of which we r supposed to select from
also we used labels to basically label the averages from the limits given min=leastpop, 25 to 50 below avg , 50 to 75 avg, 75+ popular

in above code 'pd.cut' helps in categorization .

```python
labels = ['not_popular', 'below_avg', 'average', 'popular']

categorize_col(df, 'Vote_Average', labels)

df['Vote_Average'].unique()
```
